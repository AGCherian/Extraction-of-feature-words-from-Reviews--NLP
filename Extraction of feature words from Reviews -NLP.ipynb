#### B. Compactness Prunning

```python
#get all the features that contains 2 or more words
two_word_features = [tuple(item) for item in apr_df.itemsets  if len(item) >=2]
```

```python
# keep the global initialization of this array.
close3_indexs = []

# define punctuations
puncts = list(string.punctuation)

# check wether the values are not separated by punctions or conjunctions
# if not separated by conjuctions return True
    
def chk4_ConjPunct(close3_indexs, tw_feature, review):
    
    # initialize the flag
    not_separate_by_conjORpunct = True
    
    # use the close three indexes and check for the words in between.
    for ci in close3_indexs:
        
        ci = sorted(ci)
        
        # Conjuction checks
        # get the tags of all in between words and check for 'CC'
        tags = nltk.pos_tag(review[ci[0]+1:ci[1]])
        
        for tag in tags:
            if tag[1] == 'CC':
                not_separate_by_conjORpunct = False
                break
                
        # Punctuation checks
        # Check for each of the in between words whether it is punctuaton or not.
        for word in review[ci[0]+1:ci[1]]:
            
            if word in puncts:
                
                not_separate_by_conjORpunct = False
                break
    
#     if not_separate_by_conjORpunct:
#         print(review[ci[0]:ci[1]+1])
#         print(tags)
        
    return not_separate_by_conjORpunct


# to check the distance(no.of words) between the two words.
# Also check multiple occurances of the words if present
    
def distance_tw(tw_feature, review):
    
    # Flag to check if atleast one with <=3 distance is there
    closer_flag = False 
    
    # indexes of first element
    idx0 = [i for i, x in enumerate(review) if x == tw_feature[0]]
    # indexes of second element
    idx1 = [i for i, x in enumerate(review) if x == tw_feature[1]]
    
    # check to see nouns are <= 3 word length. Save those indexes for the review. 
    for i in idx0:
        for j in idx1:
            if np.abs(i-j) <=3:
                closer_flag = True
                close3_indexs.append((i,j))
    
    return closer_flag


# to get compactness count of each of the above two word feature
compact_dict = {}
# to keep the total count of each two word feature
total_tw_dict = {}

for review in reviews:
    
    # take the review as a list
    review = word_tokenize(review)
    
    for tw_feature in two_word_features: #[('camera', 'battery')]:
        
        if all(f in review for f in tw_feature):
            
            # keep the total count of each two word feature for later
            total_tw_dict[tw_feature] = total_tw_dict.get(tw_feature,0) + 1
            
            # To save indexes of the two word pairs with word length <=3. 
            # This is useful in checking conjuctions and punctions
            close3_indexs = []
            
            # check the distance(no.of words) between the two words.
            if distance_tw(tw_feature, review):
                
                #check wether the values are not separated by punctions or conjunctions
                if chk4_ConjPunct(close3_indexs, tw_feature, review) :
                    
                    # two_word_feature is compact. add to dictionioary. 
                    # also keep the count of number of times it is compact for later.    
                    compact_dict[tw_feature] = compact_dict.get(tw_feature,0) + 1
                

# threshold m
m = 5

# initialize compact features list
compact_features = pd.DataFrame(columns=['Two-word Feature','Total count','Compact count'])

# check if the two word feature occurs in m sentences
# and it is compact in atleast ten percent of sentences (This is 2 in paper. But for example battery camera
# appeared in 114 reviews and found compact in 5. This was mainly because user didnt use comma in listing them.
# so i believe a percentage would chang this behaviour)

for tw_feature in compact_dict.keys():
    if (total_tw_dict[tw_feature] >= m) and (compact_dict[tw_feature] >= .1 * total_tw_dict[tw_feature] ):
        compact_features.loc[len(compact_features),:] = [tw_feature, total_tw_dict[tw_feature], compact_dict[tw_feature]]

print(compact_features)
```

#### C. Redundancy Pruning

```python
one_word_features = [tuple(item)[0] for item in apr_df.itemsets  if len(item) == 1]
```

```python
# p-support threshold in percentage

p_threshold = .5

redundancy_df = pd.DataFrame(columns=['One-word feature', 'Total count', 'With Other\'s count', 'p-support'])

for ow_feature in one_word_features:
    
    # Get the two word reviews with this ow_feature in it  
    # initialize two word reviews with this one word list
    tw_with_crnt_ow_list = []
    
    for tw_feature in two_word_features:
        
        if ow_feature in tw_feature:
            
            tw_with_crnt_ow_list.append(tw_feature)
    
    # initialize variable to keep the total number for each one word review
    total_ow_count = 0
    
    # get the total count where these two word features appears in reiews
    # initialize total count of these
    
    tw_with_crnt_ow_count = 0
    
    # total count of reviews with this feature
    for review in reviews:
    
        # take the review as a list
        review = word_tokenize(review)
        
        if ow_feature in review:
            
            total_ow_count += 1
                       
        # get the total count where these two word features appears in reiews
        for tw_feature in tw_with_crnt_ow_list:
        
        # check if any of these in the review then add 1
            if all(f in review for f in tw_feature):
                
                tw_with_crnt_ow_count += 1
                break
                  
    # now determine the pure support.   
    p_support = total_ow_count - tw_with_crnt_ow_count
 
    redundancy_df.loc[len(redundancy_df),:] = [ow_feature, total_ow_count, tw_with_crnt_ow_count, p_support]
```
